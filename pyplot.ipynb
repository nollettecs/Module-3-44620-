# Question 2: Random Number Chart

### - Creating a plot and scatter chart from random integers

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
# Must Show 2 Charts for Question 2

import matplotlib.pyplot as plt
from random import randint


# Generate a sequence of 5 random integers
random_numbers = [randint(1, 100) for _ in range(5)]
x_values = list(range(len(random_numbers)))  # Use indices as x-values

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

#Customize the Graph
title_font = {'fontsize': 18, 'fontweight': 'bold', 'color': 'White', 'family': 'Serif'}
plt.style.use('dark_background')


# Scatter plot
ax1.scatter(x_values, random_numbers, marker='*', color='Skyblue')
ax1.set_title("Scatter Plot of Random Numbers", fontdict=title_font)
ax1.set_xlabel("Index", {'fontsize': 14, 'fontweight': 'bold', 'color': 'White', })
ax1.set_ylabel("Random Number", {'fontsize': 14, 'fontweight': 'bold', 'color': 'White', })

# Line plot
ax2.plot(x_values, random_numbers, color='teal', marker='*', linestyle='-')
ax2.set_title("Line Plot of Random Numbers", fontdict=title_font)
ax2.set_xlabel("Index", {'fontsize': 10, 'fontweight': 'bold', 'color': 'White', })
ax2.set_ylabel("Random Number", {'fontsize': 10, 'fontweight': 'bold', 'color': 'White', })

# Show the plot
plt.show()
```

You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.
```python
#sizes = [10, 20, 50, 100, 1000, 5000]
#insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
#merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]


import matplotlib.pyplot as plt

# Data
sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

# Create a figure and axis
plt.figure(figsize=(12, 8))

# Plot Insertion Sort times as a line plot
plt.plot(sizes, insertion_sort_times, marker='o', color='Green', label='Insertion Sort', linestyle='--', linewidth=3)

# Plot Merge Sort times as a line plot
plt.plot(sizes, merge_sort_times, marker='*', linestyle='--', linewidth=3, color='Orange', label='Merge Sort',)

# Customize the graph
plt.style.use('ggplot')

# Set the title and labels
plt.title('Sorting Algorithm Execution Times', {'fontsize': 22, 'fontweight': 'ultralight', 'color': 'Black', 'family': 'cursive' } )
plt.xlabel('Size of List', fontsize=14, fontweight='heavy')
plt.ylabel('Time (seconds)', fontsize=14, fontweight='heavy')

# Add a legend
plt.legend(loc='upper left')

# Show grid for better readability
plt.grid(True)

# Show the plot
plt.xscale('log')  # Optional: change x-axis to log scale for better visualization
plt.yscale('log')  # Optional: change y-axis to log scale for better visualization
```
